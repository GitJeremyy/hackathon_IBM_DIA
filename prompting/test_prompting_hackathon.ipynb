```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model and tokenizer
model_name = "mistralai/Mistral-7B-Instruct-v0.3"  # latest instruct model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")

def ask_mistral(query, system_prompt="Vous êtes un assistant IA intelligent spécialisé dans la réponse aux questions en utilisant des informations contextuelles pertinentes dans le cadre d'un RAG."):
    prompt = f"<s>[INST] {system_prompt}\n{query} [/INST]"
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(**inputs, max_new_tokens=300)
    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return answer.split("[/INST]")[-1].strip()
```

```python
query="INSTRUCTIONS :\n1. Analysez attentivement la question posée par l'utilisateur\n2. Examinez la liste de questions les plus similaires dans notre base de questions \n3. Utilisez UNIQUEMENT les informations fournies dans le contexte pour répondre\n4. Choisissez la question la plus similaire à la question posée par l'utilisateur. Si aucune question n'est suffisament similaire, n'en sélectionnez aucune\n5. Retournez l'id de la question ou 'aucune' si aucune n'est suffisament proche\nQUESTION DE L'UTILISATEUR :\n{Que faire si j'ai un rdv médical et que j'ai eu une absence?}\nINFORMATIONS CONTEXTUELLES RÉCUPÉRÉES :\n{2133:Un certificat médical peut-il excuser une absence?\n2134:Un rendez-vous médical peut-il excuser une absence?\n2135:Quel est le quota d'absence autorisée par semestre?}\nCONSIGNES DE RÉPONSE :\n- Répondez EXCLUSIVEMENT l'id de la question similaire\n- Basez votre réponse EXCLUSIVEMENT sur les informations contextuelles fournies \n- Si vous ne trouvez pas de question suffisament similaire dans le contexte, dites 'aucune'\n- Ne répondez QUE l'id OU 'aucune'\nRÉPONSE :"

answer=ask_mistral(query=query)
print(answer)
```
